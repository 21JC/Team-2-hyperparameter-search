```python
import pandas as pd
import json
import ast
```

```python
Trail = pd.read_csv('hyperparameter_data.csv')
```

```python
Trail.head()
```

```python
# Convert str to dict
Trail['hyperparameters'] = Trail['hyperparameters'].apply(lambda x: ast.literal_eval(x))
```

```python
# Convert dict to columns
pd.DataFrame(list(Trail['hyperparameters'].values))
```

```python
# Concat the data
hyperparameter = pd.concat([Trail,pd.DataFrame(list(Trail['hyperparameters'].values))],axis=1 )
```

```python
hyperparameter.info()

# Found 2 columns with missing values
```

```python
# Get model performance from learning summaries

hyperparameter['train_result']=None
hyperparameter['test_result']=None

for i in range(len(hyperparameter)):
    file_name = 'C:\\Users\\cheng\\Desktop\\235\\data\\learning_summaries\\' + hyperparameter['trial_id'][i] + '_' + hyperparameter['parent_study_id'][i] +'.csv'
    trail = pd.read_csv(file_name)
    # Remove early stopped iterations
    trail = trail[trail.train_summaries != 0]
    # Sub-in with the end result
    hyperparameter.loc[i,['train_result']] = trail.iloc[-1].train_summaries
    hyperparameter.loc[i,['test_result']] = trail.iloc[-1].test_summaries
```

```python
# Output cleaned data
hyperparameter.loc[:,'trial_id':].to_csv('Cleaned dataset V1.csv')
```
